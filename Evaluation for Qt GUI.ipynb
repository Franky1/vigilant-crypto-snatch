```python
import pandas as pd
```

```python
trades = pd.read_json('trades.js')
trades.columns
```

```python
volume_per_asset_pair = (
    trades[['coin', 'fiat', 'volume_coin', 'volume_fiat']]
    .groupby(['coin', 'fiat'])
    .agg(count=('volume_fiat', 'count'), total_fiat=('volume_fiat', 'sum'),
         total_coin=('volume_coin', 'sum'),
         )
    .reset_index()
)
volume_per_asset_pair['average_price'] = volume_per_asset_pair['total_fiat'] / volume_per_asset_pair['total_coin']
volume_per_asset_pair = volume_per_asset_pair.rename(columns={'coin': 'Coin', 'fiat': 'Fiat', 'total_fiat': 'Total Fiat', 'total_coin': 'Total Coin', 'average_price': 'Average Price', 'count': 'Trades'})
volume_per_asset_pair
```

```python
(
    trades[['coin', 'fiat', 'trigger_name', 'volume_coin', 'volume_fiat']]
    .groupby(['coin', 'fiat'])
    .agg(total_fiat=('volume_fiat', 'sum'),
         total_coin=('volume_coin', 'sum'),
         count=('volume_fiat', 'count'))
    .reset_index()
)
```

```python
volume_per_asset_pair = (
    trades[['coin', 'fiat', 'trigger_name', 'volume_coin', 'volume_fiat']]
    .groupby(['coin', 'fiat', 'trigger_name'])
    .agg(count=('volume_fiat', 'count'), total_fiat=('volume_fiat', 'sum'),
         total_coin=('volume_coin', 'sum'),
         )
    .reset_index()
)
volume_per_asset_pair['average_price'] = volume_per_asset_pair['total_fiat'] / volume_per_asset_pair['total_coin']
volume_per_asset_pair = volume_per_asset_pair.rename(columns={'coin': 'Coin', 'fiat': 'Fiat', 'total_fiat': 'Total Fiat', 'total_coin': 'Total Coin', 'average_price': 'Average Price', 'count': 'Trades', 'trigger_name': 'Trigger Name'})
volume_per_asset_pair
```
